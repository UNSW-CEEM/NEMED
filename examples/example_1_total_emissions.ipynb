## NEM Emissions Data tool
### Example One: Total Emissions
Using the package to calculate total emissions for a historical period in the NEM

#### Import Packages

```python
from nemed.process import *
```

#### Define Local Parameters

```python
cache = "" # Change this to a existing local empty folder or leave blank "" to create new cache
start_time = "2021/12/27 00:00:00" # Start of historical period to collect data for
end_time = "2021/12/28 00:00:00" # End of historical period to collect data for
filter_units = None
filter_regions = None
```

#### Get Total Emissions by Dispatch Interval and for each DUID

```python
interval_totals = get_total_emissions_by_DI_DUID(start_time, end_time, cache, filter_units, filter_regions)
interval_totals
```

#### Get Total Emissions at a Regional level by Dispatch Interval

```python
region_stats = get_total_emissions_by_(start_time, end_time, cache, by="interval")
region_stats['Total_Emissions']
```

We can also see the total energy and emissions intensity metrics

```python
region_stats['Energy']
```

```python
region_stats['Intensity_Index']
```
